```python
#importing necessary modules
import pandas as pd
from pandas import Series, DataFrame
import pandas.io.data
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np
from scipy import stats,spatial
import math



#import csv as a dataframe and do apropriate data munging
df = pd.read_csv('http://www.cs.stonybrook.edu/~skiena/591/hw1/country-data.csv')
df = df.replace(to_replace='unknown', value=np.nan)
df = df.fillna(df.mean())
df = df.dropna()


#converting dataframe series as float type
df['Life Expectancy'] = df['Life Expectancy'].astype(float)
df['Health Expenditure/GDP'] = df['Health Expenditure/GDP'].astype(float)
df['Internet Users (circa 2009)'] = df['Internet Users (circa 2009)'].astype(float)
df['Net Migration Rate/1000 Population'] = df['Net Migration Rate/1000 Population'].astype(float)
df['Literacy Rate (%)'] = df['Literacy Rate (%)'].astype(float)
df['Labor Force'] = df['Labor Force'].astype(float)



Countries = df['Country Name']
Area = df['Area (sq km)']
Ps = df['Population']
LEs = df['Life Expectancy']
HEs = df['Health Expenditure/GDP']
IUs = df['Internet Users (circa 2009)']
NMRs = df['Net Migration Rate/1000 Population']
LRs = df['Literacy Rate (%)']
LFs = df['Labor Force']
Lats = df['Latitude of Capital']
Lons = df['Longitude of Capital']


#conversion of latitude and Longitude
Lat = []
Long = []
for lat in Lats:
    lat = lat.replace(" ",",")
    a,b,c = lat.split(",")
    if(c=='S'):
        lat = "-"+a+"."+b
    else:
        lat = a+"."+b
    lat = float(lat)
    Lat.append(lat)
    
for lon in Lons:
    lon = lon.replace(" ",",")
    a,b,c = lon.split(",")
    if(c=='W'):
        lon = "-"+a+"."+b
    else:
        lon = a+"."+b
    lon = float(lon)
    Long.append(lon)
    
```

```python

fig1 = plt.scatter(LEs,HEs, s=100)
fig1_title = plt.title("Figure:1")
fig1_xl = plt.xlabel("Life Expectancy")
fig1_yl = plt.ylabel("Health Expenditure/GDP")
plt.grid(True)

```
